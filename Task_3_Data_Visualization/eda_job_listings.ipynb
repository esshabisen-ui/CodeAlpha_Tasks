## Task 3: DATA VISUALIZATION

Objective of DATA VISUALIZATION:
The objective is to visually represent job market trends and insights extracted from the dataset to support better decision - making.

### Chart 1: Top Job Titles by Number of Listings

```python
top_jobs = df['Job_Title'].value_counts().head(10)


plt.figure(figsize=(10,5))
sns.barplot(x=top_jobs.values, y=top_jobs.index)
plt.title("Top 10 Job Titles by Number of Listings")
plt.xlabel("Number of Job Postings")
plt.ylabel("Job Title")
plt.show()
```

**Insight**:
This chart shows the most frequently posted job roles in the dataset.  
Roles related to development and engineering appear most often, indicating high demand in the job market.

### Chart 2: Top Job Locations by Number of Listings

```python
top_locations = df['Location'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_locations.values, y=top_locations.index)
plt.title("Top 10 Job Locations by Number of Listings")
plt.xlabel("Number of Job Postings")
plt.ylabel("Location")
plt.show()
```

**Insight**:
This chart highlights the locations with the highest number of job listings.  
Certain cities appear more frequently, suggesting higher employment opportunities in those regions.

### Chart 3: Salary Distribution of Job Listings

```python
salary_counts = df['Salary'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=salary_counts.index, y=salary_counts.values)
plt.title("Distribution of Salary Ranges")
plt.xlabel("Salary Range")
plt.ylabel("Number of Job Listings")
plt.show()
```

**Insight**:
This chart shows the distribution of salary ranges across job listings.  
Mid-range salaries appear most frequently, indicating common compensation levels in the market.

### Chart 4: Job Postings Over Time

```python
df['Posted Date'] = pd.to_datetime(df['Posted Date'], dayfirst= True, errors='coerce')

date_counts = df['Posted Date'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.plot(date_counts.index, date_counts.values, marker='o')
plt.title("Job Postings Over Time")
plt.xlabel("Posted Date")
plt.ylabel("Number of Job Listings")
plt.xticks(rotation=45)
plt.show()
```

**Insight**:
This line chart represents the trend of job postings over time.  
An increase in recent dates indicates active hiring trends.

### Chart 5: Top Hiring Companies

```python
top_companies = df['Company'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_companies.values, y=top_companies.index)
plt.title("Top 10 Companies by Job Listings")
plt.xlabel("Number of Job Listings")
plt.ylabel("Company")
plt.show()
```

**Insight**:
This chart identifies companies with the highest number of job postings,  
highlighting organizations with active recruitment efforts.
